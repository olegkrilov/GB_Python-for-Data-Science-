```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
DATASET_SRC = './creditcard.csv'
MAIN_PROP = 'Class'
SEC_PROP = 'V1'
```

```python
data = pd.read_csv(DATASET_SRC)
groupped_by_class = data.value_counts([MAIN_PROP])
```

```python
figure, sections = plt.subplots(nrows=1, ncols=2)

figure.set_size_inches(8, 6)
figure.subplots_adjust(wspace=0.3)

for index in range(2):
    section = sections[index] 
    section.title.set_text(f'{"Linear" if not index else "Logariphmic"} representation')
    sections[index].bar([0, 1], groupped_by_class, log=index)
    
plt.show()
```

```python

for index, color in enumerate(['#ccc', '#f00']):
    plt.hist(data[data[MAIN_PROP] == index][SEC_PROP], color=color, density=True, alpha=.5)

plt.xlabel('Class')
plt.legend(labels=['Legal operations (Class 0)', 'Fraud operations (Class 1)'])
plt.show()
```
