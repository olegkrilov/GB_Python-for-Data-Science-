```python
import sys; sys.path.insert(0, '..')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

from lib.Core import (
    TEST_RANGE, RANDOM_SEED, PRICE,
    load_data, train_model
)
```

```python
RANDOM_FOREST_REGRESSOR = 'RandomForestRegressor'
LINEAR_REGRESSION = 'LinearRegression'

ESTIMATORS = 1000
DEPTH = 12
```

```python
X, y = load_data()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_RANGE, random_state=RANDOM_SEED)
```

```python
# Test with RandomForestRegressor
model = RandomForestRegressor(n_estimators=ESTIMATORS, max_depth=DEPTH, random_state=RANDOM_SEED)

train_results_A, score_A = train_model(
    X_train,
    y_train.values[:, 0],
    X_test,
    y_test[PRICE],
    model=model
)
```

```python
# Test with LinearRegression
train_results_B, score_B = train_model(
    X_train,
    y_train,
    X_test,
    y_test[PRICE]
)
```

```python
print(f'{RANDOM_FOREST_REGRESSOR} Score : {score_A}')
print(f'{LINEAR_REGRESSION} Score      : {score_B}')
print(f'Better result with          : {RANDOM_FOREST_REGRESSOR if score_A > score_B else LINEAR_REGRESSION}')
```
